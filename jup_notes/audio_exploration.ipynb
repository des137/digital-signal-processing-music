```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import librosa
import librosa.display
import IPython.display as ipd
```

```python
MUSIC_PATH = '../data/data/genres_original/'

def get_sample_audio(genre='blues'):
    rnd_song = random.choice(os.listdir(MUSIC_PATH + genre))
    print('Song selected:', rnd_song)
    return MUSIC_PATH + genre + '/' + rnd_song
```

```python
aud_file = get_sample_audio()
```

```python
data, sampling_rate = librosa.load(aud_file)
ipd.Audio(data, rate=sampling_rate/1)
```

```python
ipd.Audio(data, rate=sampling_rate/5)
```
